```python
#https://stackoverflow.com/questions/50239066/efficiently-standardizing-images-in-a-numpy-array
def normalize_meanstd(a, axis=None): 
    # axis param denotes axes along which mean & std reductions are to be performed
    mean = np.mean(a, axis=axis, keepdims=True)
    std = np.sqrt(((a - mean)**2).mean(axis=axis, keepdims=True))
    return (a - mean) / std

#Used code from link below to vectorize computation and avoid looping
#https://stackoverflow.com/questions/30143417/computing-the-correlation-coefficient-between-two-multi-dimensional-arrays
def corr2_coeff(A,B):
    # Rowwise mean of input arrays & subtract from input arrays themeselves
    A_mA = A - A.mean(1)[:,None]
    B_mB = B - B.mean(1)[:,None]

    # Sum of squares across rows
    ssA = (A_mA**2).sum(1);
    ssB = (B_mB**2).sum(1);

    # Finally get corr coeff
    return np.dot(A_mA,B_mB.T)/np.sqrt(np.dot(ssA[:,None],ssB[None]))
def corr_space(input_img_series, window_size=2, indices=None):
    """
    Takes an input with shape (x,y,z) where z represents the number of images in observation
    Then for the time-series of each pixel we calculate it's correlation to 
    the time series of pixels in a window in the center of the image
    
    window_size explanation, let c be the center pixel of the window and s the window_size
    it adds padding of size s to the center pixel, so that if s=2 the resulting window is the
    following:
    
    ppppp
    ppcpp
    ppppp
    
    that means, a window of 5x5, or (2s + 1)^2
    """
    x, y, z = input_img_series.shape
    center_window_pixel = x*y//2 + 1 #flattened image
    i, j = center_window_pixel // x, center_window_pixel % y
    if not indices:
        window = input_img_series[i-window_size:i+window_size+1, j-window_size:j+window_size+1,:]
    else:
        new_size = int(np.sqrt(len(indices)))
        window = input_img_series.reshape(x*y, z)[indices].reshape(new_size,new_size,z)
    #Another option is to give the indices of the window pixels
    wx, wy, wz = window.shape

    return corr2_coeff(input_img_series.reshape(x*y, z),
                     window.reshape(wx*wy, wz)).reshape(x,y,wx*wy)
```

```python
def basic_deep_hits(input_shape=(48,21,21), lr=0.001, decay=1e-6):

    model = Sequential ()
    model.add(ZeroPadding2D((3, 3), input_shape = input_shape))
    model.add(Convolution2D(32, (4, 4)))
    model.add(Activation('relu'))
    model.add(ZeroPadding2D((1, 1)))
    model.add(Convolution2D(32, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size =(2, 2)))

    model.add(ZeroPadding2D((1, 1)))
    model.add(Convolution2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(ZeroPadding2D((1, 1)))
    model.add(Convolution2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(ZeroPadding2D((1, 1)))
    model.add(Convolution2D(64, (3, 3)))
    model.add(Activation('relu'))
    model.add(MaxPooling2D(pool_size =(2, 2)))

    model.add(Flatten())

    model.add(Dense(64))
    model.add(Dropout(0.5))
    model.add(Activation('relu'))
    model.add(Dense(64))
    model.add(Dropout(0.5))
    model.add(Activation('relu'))
    model.add(Dense(2))
    model.add(Activation('softmax'))
    
    sgd = keras.optimizers.SGD(lr=lr, decay=decay)
    model.compile(loss='mean_squared_error',
                  optimizer=sgd)
    return model
```

```python
n_train = int(len(sample_images) * 0.8)
x_train, y_train = sample_images[:n_train], sample_labels[:n_train]
x_test, y_test = sample_images[n_train:], sample_labels[n_train:]

x_train = normalize_meanstd(x_train, (2,3))
x_test = normalize_meanstd(x_test, (2,3))
```

```python
#cant remember what this cell used to do
indices_max_diff = np.argmax((x_train[:][0] - x_train[:]).sum(axis=2).sum(axis=2),axis=1)
differences = np.zeros((x_train.shape[0],21,21), dtype=np.float32)
for i in range(indices_max_diff.shape[0]):
    differences[i] = (x_train[i][0] - x_train[i][indices_max_diff[i]])
true_train = np.stack((x_train[:,0], x_train[np.arange(len(x_train)), indices_max_diff], differences), axis=1)
```

```python
print(x_train.shape)
#print(x_train[np.arange(len(x_train)), indices_max_diff].shape)
#print(differences.shape)
top_k = 16
#find top_k maximum intensity pixels in first image of the series
#indices = np.argpartition(np.moveaxis(x_train, 1, 3)[0][:,:,0].ravel(), -top_k)[-top_k:]
#, sorted(np.argpartition(img[:,:,0].ravel(), -top_k)[-top_k:])
new_x_train = np.asarray([np.moveaxis(corr_space(img, 1), -1, 0) for img in np.moveaxis(x_train,1, 3)])
new_x_train_25 = np.asarray([np.moveaxis(corr_space(img, 2), -1, 0) for img in np.moveaxis(x_train,1, 3)])
new_x_test = np.asarray([np.moveaxis(corr_space(img, 1), -1, 0) for img in np.moveaxis(x_test,1, 3)])
new_x_test_25 = np.asarray([np.moveaxis(corr_space(img, 2), -1, 0) for img in np.moveaxis(x_test,1, 3)])
```

```python
batch_size = 32
epochs = 75
n_models = 10
x_epochs = np.arange(1, epochs+1)
```

# Correlation space input using a square 3x3 pixel window

```python
corr_models = [basic_deep_hits(input_shape=(9,21,21), lr=0.01) for i in range(n_models)]
history_corr_models = [model.fit(new_x_train, y_train, batch_size=batch_size, 
                        epochs=epochs, validation_split=0.2) for model in corr_models]
```

```python
corr_losses = np.asarray([history.history['loss'] for history in history_corr_models])
corr_val_losses = np.asarray([history.history['val_loss'] for history in history_corr_models])

corr_losses_mean = corr_losses.mean(axis=0)
corr_losses_std = corr_losses.std(axis=0)

corr_val_losses_mean = corr_val_losses.mean(axis=0)
corr_val_losses_std = corr_val_losses.std(axis=0)
```

```python
plt.errorbar(x_epochs, corr_losses_mean, corr_losses_std, label='training', capsize=2, linewidth=0.5, alpha=1)
plt.errorbar(x_epochs, corr_val_losses_mean, corr_val_losses_std, label='validation', capsize=2, linewidth=1, alpha=0.75)

plt.title('Training with correlation space input, 9 pixels window')
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(loc='best')
plt.tight_layout()
plt.savefig('training_corrspace_9pixels.png', dpi=300)
plt.show()
```

```python
pbb_threshold = 0.5

corr_accs = np.zeros(n_models)
corr_recalls = np.zeros(n_models)
corr_f1s = np.zeros(n_models)
corr_classified = np.zeros(n_models)
corr_predicted = []

for i, model in enumerate(corr_models):
    Y_pred_corr = model.predict (new_x_test)
    pred_labels = np.array((Y_pred_corr[:, 1] > pbb_threshold), dtype = "int")
    corr_predicted.append(pred_labels)
    corr_accs[i] = accuracy_score(y_pred=pred_labels, y_true=y_test[:,1])
    corr_recalls[i] = recall_score(y_pred=pred_labels, y_true=y_test[:,1])
    corr_f1s[i] = f1_score(y_pred=pred_labels, y_true=y_test[:,1])
    corr_classified[i] = (pred_labels == y_test[:, 1].astype(int)).sum()
```

```python
print('Accuracy: {} +- {}'.format(corr_accs.mean(), corr_accs.std()))
print('Recalls: {} +- {}'.format(corr_recalls.mean(), corr_recalls.std()))
print('F1 Score: {} +- {}'.format(corr_f1s.mean(), corr_f1s.std()))
print('Correctly classified {} out of {}'.format(corr_classified, new_x_test.shape[0]))
print('Mean correct classified: {} +- {}'.format(corr_classified.mean(), corr_classified.std()))
```
